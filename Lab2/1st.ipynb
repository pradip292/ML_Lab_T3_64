```python
# Get dataset
df_sal = pd.read_csv('Student_Marks.csv')
df_sal.head()
```

```python
# Describe data
df_sal.describe()
```

```python
# Data distribution
plt.title('Student Study Time related to Marks')
sns.distplot(df_sal['Marks'])
plt.show()
```

```python
# Data distribution
plt.title('Student Study Time related to Marks')
sns.distplot(df_sal['time_study'])
plt.show()
```

```python
# Relationship between Study time and Marks obtained by student
plt.scatter(df_sal['time_study'], df_sal['Marks'], color = 'lightcoral')
plt.title('Study time vs Marks Obtained')
plt.xlabel('Study Time')
plt.ylabel('Marks Obtained')
plt.box(False)
plt.show()
```

```python
# Splitting variables
X = df_sal.iloc[:, :1]  # independent
y = df_sal.iloc[:, 1:]  # dependent
```

```python
# Splitting dataset into test/train
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
# Regressor model
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# Prediction result
y_pred_test = regressor.predict(X_test)     # predicted value of y_test
y_pred_train = regressor.predict(X_train)   # predicted value of y_train
```

```python
# Regressor coefficients and intercept
print(f'Coefficient: {regressor.coef_}')
print(f'Intercept: {regressor.intercept_}')
```
